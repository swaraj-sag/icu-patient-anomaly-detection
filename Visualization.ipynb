```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("iris")
```

```python
df
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
df.head(1)
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df,ci=68) #Confidence Interval
plt.show()
```

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=df,hue='species',ci=None)
```

```python
sns.lineplot(x='sepal_width',y='species',data=df,hue='species',ci=None)
```

# Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,hue='species')
plt.title("Scatter Plot")
plt.show()
```

# Bar Chart

```python
sns.barplot(x='species',y='sepal_length',data=df)
plt.show()
```

# Count plot

```python
sns.countplot(x='species',data=df)
```

# Box Plot

```python
sns.boxplot(x='sepal_width',data=df)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_width',data=df)
```

# Violin Plot

```python
sns.violinplot(x='species',y='sepal_width',data=df)
```

# Strip Plot

```python
sns.stripplot(x='species',y='petal_width',data=df)
```

# Histogram Plot

```python
sns.histplot(x='petal_width',data=df)
```

```python
sns.histplot(x='petal_width',data=df,hue='species',kde=True)
```

# Dist Plot

```python
sns.distplot(df['petal_width'],color='g')
```

# Pair Plot for multi variate only for Numric columns

```python
sns.pairplot(data=df,hue='species')
```
